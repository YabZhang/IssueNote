```python
# constructor
a = np.array([0, 1, 2 ,3])  # 1-D
b = np.array([[0, 1, 2], [3, 4, 5]])  # 2-D
print('a: ', a)
print('b: ', b)
```

```python
print(b.ndim) # dimension
print(b.shape) # type
print(b.reshape((3, 2))) # change type
```

```python
# func of creating arrays
print(np.arange(10)) # 0 .. n - 1, like range
print(np.arange(1, 9, 2))
```

```python
print(np.linspace(0, 1, 6)) # start, end, num-points, including endpoint
print(np.linspace(0, 1, 5, endpoint=False)) # no end-point
```

```python
# common
print(np.ones((3, 3))) 
print(np.zeros((2, 2)))
print(np.eye(3))
print(np.diag(np.array([1, 2, 3, 4])))  # diagnal matrix
```

```python
# random
print(np.random.rand(4))  # uniform in [0, 1]
print(np.random.randn(4))  # Gaussian
print(np.random.seed(1234)) # set random seed
```

```python
a = np.array([1, 2, 3])
print(a.dtype)  # dtype('int64')

b = np.array([1., 2., 3.])
print(b.dtype)  # dtype('float64')

a = np.ones((3, 3))
print(a.dtype)  # default float64
```

```python
d = np.array([1+2j, 3+4j, 5+6j])
print(d.dtype)

e = np.array([True, False, False, True])
print(e.dtype)

f = np.array(['Bonjour', 'Hello', 'Hallo'])
print(f.dtype)
```

```python
a = np.arange(10)
b = a[::2]
print(a, b)
print(np.may_share_memory(a, b))
b[0] = 12
print(a)

# copy
a = np.arange(10)
c = a[::2].copy()  # create a duplicate of `a` named as 'c'
c[0] = 12
print(a) # a has no change
```
